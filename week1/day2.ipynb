Connecting local dowload Ollama with code

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown,display
```

```python
OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {"Content-Type": "application/json"}
MODEL = "llama3.2"
```

```python
# Create a messages list using the same format that we used for OpenAI
messages = [
    {"role":"user","content":"Provide some of the real time  application of Generative AI description"}
]
```

```python
payload = {
    "model":MODEL,
    "messages":messages,
    "stream" : False
}
```

```python
# make sure model is loaded

!ollama pull llama3.2
```

```python
# If this doesn't work for any reason, try the 2 versions in the following cells
# And  check the instructions page of ollama
response  = requests.post(OLLAMA_API,json= payload,headers=HEADERS)
print(response.json()['message']['content'])
```

We can do above using Ollama Python package

```python
import ollama # import ollama package
response = ollama.chat(model = MODEL, messages = messages) # use as direct chat
print(response['message']['content'])
```

# Using openAI python library to connect to Ollama

```python
# We can use OpenAI client library to call ollama
from openai import OpenAI
ollama_via_openai = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")

response = ollama_via_openai.chat.completions.create(
    model = MODEL,
    messages=messages
)

print(response.choices[0].message.content)
```
